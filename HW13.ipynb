```python
# Define parameters etc.

dx = 0.002           # step size in space
dt = 0.002           # step size in time

x_min = 0
x_max = 4
t_max = 7

x = np.arange(x_min, x_max+dx, dx)
t = np.arange(0, t_max+dt, dt)

#initial condition
def initial_condition(x):
    return np.where((x > 0) & (x < 1), 1, 0)

#speed function
def speed_function(u):
    return u**2/2

#exact solution
def exact_sol(x,t):
    
    solultion = np.zeros_like(x)
    
    if t <=2 :
        
        for i in range(len(x)):
        
            if x[i] <= 0:

                solultion[i] =  0

            elif x[i]>=0 and x[i]<= t:

                solultion[i] = x[i]/t

            elif x[i]>t and x[i]< 1+t/2:

                solultion[i] = 1

            else:

                solultion[i] = 0
        
    else:
        
        for i in range(len(x)):
        
            if x[i] < 0:

                solultion[i] = 0

            elif x[i]>=0 and x[i] <= np.sqrt(2*t):

                solultion[i] = x[i]/t

            else:

                solultion[i] = 0
    
    return solultion
```

```python
# Exact Solution
scatter_plot_1(x,exact_sol)
```
