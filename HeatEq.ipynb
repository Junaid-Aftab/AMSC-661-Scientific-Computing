```python
import numpy as np
import matplotlib.pyplot as plt
import math
import seaborn
plt.style.use('seaborn')
```

```python
def func(x,t,N):
    
    u = 0
    
    for n in range(N):
        
        u = u + ((8*(-1)**n)/(np.pi*(2*n+1)**2))*np.sin((n+1/2)*x)*np.exp(- t*(2*n+1)**2/4 )
        
    return u

T = [0,2]
M = [1,2,5,10,100]

x = np.linspace(0, np.pi, 100)
u = func(x,1,100)

# scatter plots

fig, axs = plt.subplots(nrows=len(T), ncols=len(M), figsize=(15, 7))

for i in range(len(T)):
    
    for j in range(len(M)):
        
        axs[i,j].plot(x, func(x,T[i],M[j]))
        #axs[i,j].plot(x, x)
        axs[i, j].set_title('T =  {} '.format(T[i]) + ' M =  {}'.format(M[j]))
        
plt.tight_layout()
plt.show()
```
